```python
cam = cv2.VideoCapture(0)
cam.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
cam.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)
```

```python
# 從攝影機擷取一張影像
ret, frame = cam.read()

# 顯示圖片
plt.imshow(frame)

# 若按下 q 鍵則離開迴圈

```
